### Verification

```python
from cmath import cos, sin, pi
from math import log2

def re_arange(a): # len(a) = 2^k, where k \in Z
    # there is a more efficient way to implement this using bit flipping
    if len(a) <= 2:
        return a
    b = [a[i] for i in range(0, len(a), 2)]
    c = [a[i+1] for i in range(0, len(a), 2)]
    return re_arange(b) + re_arange(c)

def omega(n, k):
    return complex(cos(2*pi*k/n), sin(2*pi*k/n))

def _FFT_Iter(arr, inv = False):
    n = len(arr)#end - start # length, shouldbe 2^k
    result = [None] * n 
    if n == 0:
        return
    for k in range(n):
        k_prime = k if k < n//2 else k - n//2
        k_prime = -1 * k_prime if inv else k_prime
        if k < n//2:
            result[k] = arr[k] + omega(n, k_prime) * arr[k+n//2]
        else:
            result[k] = arr[k-n//2] - omega(n, k_prime) * arr[k] 
    return result


def _FFT(arr, inv = False, verbose = True):
    n = len(arr)
    arr = re_arange(arr)
    if verbose:
        print("after re-order:{}\n".format(arr))
    n_iter = int(log2(n))
    for i in range(1, n_iter+1):  # log(n)
        sub_group_len = pow(2, i)
        n_sub_group = n // sub_group_len
        for j in range(n_sub_group): 
            start, end = j*sub_group_len, (j+1)*sub_group_len
            sub_group = arr[start : end]
            arr[start : end] = _FFT_Iter(sub_group, inv)
        if verbose:    
            print("Iter{}: {}\n".format(i, arr))   
    return arr

        
def FFT(arr, verbose = True):
    return _FFT(arr, inv = False, verbose = verbose)

        
def IFFT(arr, verbose = True):
    a = _FFT(arr, inv = True, verbose = verbose)  
    n = len(a)
    a_over_n = [a[i]/n for i in range(n)]
    return a_over_n
```

```python
a = [1, -3, 0, 1, 0, 0, 0, 0]
b = [2, -1, 0, 1, 0, 0, 0, 0]
```

```python
A, B = FFT(a), FFT(b)
```

```python
C = [A[i]*B[i] for i in range(len(A))]
print(C)
```

```python
c = IFFT(C)
print(c)
```
