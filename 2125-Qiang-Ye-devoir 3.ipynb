**Question 1**

Provez que $f \in \Theta(g)$ si, et seulement si  $g\in\Theta(f)$, si, et seulement si $\Theta (f) = \Theta (g)$.


**Answer 1**
1. Recall the justification of $f \in O(g) \Leftrightarrow g \in \Omega(f)$:
\begin{align}
f\in O(g) &\Leftrightarrow \exists n_0 \ge 1, c_0 \ge 0, \forall n > n_0, f(n) \le c_0 g(n) \\
&\Leftrightarrow \exists n_0 \ge 1, c_0 \ge 0, \forall n > n_0, g(n) \ge \frac{1}{c_0} f(n)\\ 
&\Leftrightarrow \exists n_0 \ge 1, c_1 = \frac{1}{c_0} \ge 0, \forall n > n_0,  g(n) \ge c_1 f(n)\\
&\Leftrightarrow g\in \Omega(f)\\
\end{align}

Recall the justification of $f \in \Theta(f)$:

$\exists n_0 = 1, c_0 = 1, c_0' = 1, \forall n \ge n_0, c_0 f(n) \le f(n) \le c_0' f(n)$.
Obviously, it is true that: $$f \in \Theta(f) \tag{1.0}$$.

**Part1**. Prove that $g\in \Theta(f) \Leftrightarrow f\in \Theta(g)$.

\begin{align}
g\in \Theta(f) &\Leftrightarrow g\in O(f) \land g\in \Omega(f)\\ 
&\Leftrightarrow f\in \Omega(g) \land f\in O(g)\\
&\Leftrightarrow f\in \Theta(g)\\
\end{align}

**Part2**. Prove that $g\in \Theta(f) \Leftrightarrow \Theta(f) = \Theta(g)$.

First, we have:
\begin{align}
\Theta(f) = \Theta(g) &\Leftrightarrow (\forall h \in \Theta(f) \Rightarrow h \in \Theta(g)) \land (\forall h\in \Theta(g)\Rightarrow h\in \Theta(f))\\
\end{align}
and
\begin{align}
f\in \Theta(g) &\Leftrightarrow f \in O(g) \land f \in \Omega(g)\\
&\Leftrightarrow \exists n_0 \ge 1, c_0 \ge 0, c_0' \ge 0, \forall n\ge n_0, c_0, c_0',\  c_0 g(n) \le f(n) \le c_0' g(n) \\
&\Leftrightarrow \exists n_0 \ge 1, c_0 \ge 0, c_0' \ge 0, \forall n\ge n_0, c_0, c_0',\  g(n) \le \frac{1}{c_0} f(n) \land \frac{1}{c_0'} f(n) \le g(n) \tag{1.1}\\
\end{align}

Suppose any function $h:\mathbb N \to \mathbb R^{\ge0} \in \Theta(g)$:

\begin{align}
h\in \Theta(g) &\Leftrightarrow h \in O(g) \land h \in \Omega(g)\\
&\Leftrightarrow \exists n_1 \ge 1, c_1 \ge 0, c_1' \ge 0, \forall n\ge n_1, c_1 g(n) \le h(n) \le c_1' g(n) \tag{1.2}\\
\end{align}

Based on formular (1.1),(1.2), we have:

$\exists n_2 = \max\{n_0, n_1\} \ge 1, c_0 \ge 0, c_0' \ge0, c_1\ge 0, \text{ and } c_1' \ge0$,

$\forall n\ge n_2, \frac{c_1}{c_0'} f(n) \le h(n) \le \frac{c_1'}{c_0} f(n)$.

That is, $$h \in \Theta(g) \Rightarrow h \in \Theta(f) \tag{1.3}$$

Similarly, we also have: 
$$ h \in \Theta(f) \Rightarrow h \in \Theta(g) \tag{1.4}$$


Based on (1.1),(1.3),(1.4), we have, $$ f\in \Theta(g) \Rightarrow \Theta(f) = \Theta(g) \tag{1.5}$$
<br>

Based on (1.0), we have:
\begin{align}
\Theta(f) = \Theta(g) &\Rightarrow f\in \Theta(f) \\
&\Rightarrow f\in \Theta(g) \tag{1.6}\\
\end{align}

Finally, based on (1.5) and (1.6), we proved:
\begin{align}
f \in \Theta(g) &\Leftrightarrow \Theta(f) = \Theta(g)\\
\end{align}

\begin{aligned}
 -1 \le  \sin(n) \le 1 
 &\Rightarrow -n \le  n\sin(n) \le n \\ 
 &\Rightarrow 2n-n \le 2n + n\sin(n) \le 2n+n \\

\end{aligned}

**Question3**

Prouvez que $f(n) = \lg n$, $p(n) = \sum_{i=0}^{k}a_in^{k-i}, a_i\in \mathbb R, i,k \in \mathbb N$, sont lisses.


**Answer3**

**part1** prove that $f(n) = \lg n$ is lisse.

First, $\exists n_0 = 1, \forall n \ge n_0$
\begin{align}
f(n+1) - f(n) &= \lg(n+1) - \lg n \\
&=  \lg \frac{n+1}{n} \\
&= \lg(1+\frac{1}{n}) \\
&\ge 0 \tag{3.1}\\
\end{align}
We can say, function $f(n) = \lg n$ is **e.n.d.** (éventuellement non-décroissante) 

Let $b\in \mathbb N^{\ge 2}, c_b \in \mathbb R^{> 1} \subset \mathbb R^{\ge 0} \$,

\begin{align}
f(bn) - c_b f(n) \le 0 &\Leftrightarrow \lg(bn) - c_b \lg n \le 0 \\
&\Leftrightarrow \lg n + \lg b - c_b\lg n \le 0\\
&\Leftrightarrow (1 - c_b) \lg n + \lg b \le 0\\
&\Leftrightarrow \lg n \ge  \frac{\lg b}{c_b - 1}\\
&\Leftrightarrow n \ge  2^{\frac{\lg b}{c_b - 1}}  \\
\end{align}
Which means,

$\exists c_b \in \mathbb R^{> 1} \subset \mathbb R^{\ge 0}, n_b = \lceil 2^{\frac{\lg b}{c_b - 1}} \rceil \in \mathbb N, , \forall n \ge n_b, \forall b \in \mathbb N^{\ge 2}$,

$$f(bn) \le c_b f(n) \tag{3.2}$$

Based on (3.1) and (3.2) We can say: $\lg n$ is lisse.


<br>


**part2** prove that $p(n) = \sum_{i=0}^{k}a_i n^{k-i}, a_i\in \mathbb R, i,k \in \mathbb N$ is lisse.
<hr>
To prove $p(n) = \sum_{i=0}^{k} a_i n^{k-i}, a_i \in \mathbb R, i,k \in \mathbb N$ is lisse,

we **need more constrain for $a_i$**: $a_0 > 0$, which is a special case of more general constrains:$$ a_j > 0 \text{ and }\forall i<j, a_i = 0, \text{ where }j < k$$
Otherwise, $p(n)$ may not always **e.n.d.** or lisse.

We know that, for $x, r_i \in \mathbb R, a_0 \prod_{i=0}^{k}(x-r_i)$ can be expanded to the form of $\sum_{i=0}^{k} a_i x^{k-i}$, that is:
$$\prod_{i=0}^{k}(x-r_i) = \frac{1}{a_0} \sum_{i=0}^{k} a_i x^{k-i}$$
where $a_k = a_0 (-1)^k \prod_{i=0}^{k} |r_i|$.

So, $\exists C \in \mathbb R$, satisfies:

$$\sum_{i=0}^{k} a_i x^{k-i} = a_0 \prod_{i=0}^{k}(x-r_i) + C$$

As $n\in \mathbb N \subset \mathbb R$, it is true that:

$$p(n) = \sum_{i=0}^{k} a_i n^{k-i} = a_0 \prod_{i=0}^{k}(n-r_i) + C \tag{3.3}$$

$\exists n_0 = max\{r_i \ |\ i \in \mathbb N, i \le k\}$, such that: $\forall n \ge n_0, n-r_i \ge 0$, and:
\begin{align}
p(n+1) - p(n) &= a_0\prod_{i=0}^{k}(n+1-r_i) + C - \Bigl( a_0\prod_{i=0}^{k}(n-r_i) + C \Bigr) \\
&= a_0\prod_{i=0}^{k}(n+1-r_i) - a_0\prod_{i=0}^{k}(n-r_i) \\
&\ge a_0\prod_{i=0}^{k}(n-r_i) - a_0\prod_{i=0}^{k}(n-r_i) \\
&= 0 \\
\end{align}
We can say, function $p(n)$ is **e.n.d.** (éventuellement non-décroissante) 


Let $$n_b = \lceil max\{r_i,\ r_i\Bigl(1+\frac{1}{b} \Bigr) \ |\  i \in \mathbb N, i \le k\} \rceil \tag{3.4}$$
such that $\forall n \ge n_b$, we have:
$$n-r_i \ge 0,\ bn-r_i \ge 0,\ p(bn)>0,\ p(n)>0 \tag{3.5}$$, and,
$$r_i\Bigl(1+\frac{1}{b} \Bigr) \le n, \text{ where } 0 \le i \le k \tag{3.6}$$

let $b \in \mathbb N^{\ge 2}, c_b = b^{2k}$, such that: $c_b \in \mathbb R^{>0}$.

Based on (3.5), we have:

\begin{align}
p(bn) - c_b p(n) <= 0 &\Leftrightarrow \frac{p(bn)}{c_b p(n)} \le 1\\
&\Leftrightarrow \frac{\prod_{i=0}^{k}(bn-r_i) + C}{c_b \prod_{i=0}^{k}(n-r_i) + C} \le 1\\
&\Leftrightarrow \frac{\prod_{i=0}^{k}(bn-r_i)}{b^{2k} \prod_{i=0}^{k}(n-r_i)} \le 1\\
&\Leftrightarrow \frac{\prod_{i=0}^{k}(bn-r_i)}{\prod_{i=0}^{k}b^{2}(n-r_i)} \le 1  \tag{3.7}\\
\end{align} 

To prove (3.7), we only need to prove $\frac{bn-r_i}{b^2(n-r_i)} \le 1,$ for all $i$ where $ 0 \le i \le k$:
\begin{align}
\frac{bn-r_i}{b^2(n-r_i)} \le 1 &\Leftrightarrow bn-r_i \le b^2(n-r_i)\\
&\Leftrightarrow bn-r_i \le b^2 n - b^2 r_i \\
&\Leftrightarrow (b^2-1)r_i \le (b^2-b)n\\
&\Leftrightarrow (b+1)(b-1)r_i \le b(b-1)n\\
&\Leftrightarrow  \frac{b+1}{b} r_i \le n\\
&\Leftrightarrow  \Bigl(1+\frac{1}{b} \Bigr) r_i\le n \tag{3.8}\\
\end{align}

(3.8) is same with (3.6),which is implied by (3.4), the choose of $n_b$.

Finally, we can say $p(n) = \sum_{i=0}^{k} a_i n^{k-i}, a_i \in \mathbb R, a_0 >0, i,k \in \mathbb N$ is lisse.
