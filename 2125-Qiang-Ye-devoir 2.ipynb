**Question**

Prouvez que $\text{lg}(n!) \in \Theta(n\text{lg}n)$. Rappel: $\text{lg}x = \text{log}_{2}x$.
<hr>

**Answer**

Soit functionnes $f,\ g:\mathbb{N} \to \mathbb{R^{\ge 0}}$.

On a,
$$f(x) \in \Theta(g(x)) \Leftrightarrow f(x) \in O(g(x)) \land f(x) \in \Omega(g(x))$$

Alors, 
$$\text{lg}(n!) \in \Theta(n\text{lg}n) \Leftrightarrow \text{lg}(n!) \in O(n\text{lg}n) \land \text{lg}(n!) \in \Omega(n\text{lg}n)$$

**Partie 1**: prouver que $\text{lg}(n!) \in O(n\text{lg}n)$

\begin{aligned}
\text{lg}(n!) &= \text{lg}\bigl(\prod_{i = 1}^{n} i\bigr) \\
&= \text{lg}\bigl(n \cdot (n-1) \cdot (n-2) \cdot \cdots \cdot 2 \cdot 1\bigr) \\
&\le \text{lg}(n \cdot n \cdot n \cdot \cdots \cdot n \cdot n ) \\
&= \text{lg}(n^n)\\
&= n\text{lg}n \\
\end{aligned}
$\exists\ n_0 = 1, c = 1$, tel que: $\forall n\ge n_0$, on a,
$$\text{lg}(n!) \le c\ n\text{lg}n $$
On peut dire,
$$\text{lg}(n!) \in O(n\text{lg}n)$$

**Partie 2**: prouver que $\text{lg}(n!) \in \Omega(n\text{lg}n)$
\begin{aligned}
\text{lg}(n!) &= \text{lg}(1 \cdot 2 \cdot 3 \cdot \cdots \lceil \frac{n}{2} \rceil \cdot \cdots (n-1) \cdot n)\\
&= \text{lg}1 + \text{lg}2 + \cdots + \text{lg}\lceil \frac{n}{2} \rceil + \cdots + \text{lg}(n-1) + \text{lg}n \\
&\ge 0 + 0 + \cdots + 0 + \cdots + \text{lg}\lceil \frac{n}{2} \rceil + \text{lg}(\lceil \frac{n}{2} \rceil +1) + \cdots + \text{lg}(n-1) + \text{lg}n\\
&= \text{lg}\lceil \frac{n}{2} \rceil + \text{lg}(\lceil \frac{n}{2} \rceil +1) + \cdots + \text{lg}(n-1) + \text{lg}n\\
&\ge \text{lg}\lceil \frac{n}{2} \rceil + \text{lg}\lceil \frac{n}{2} \rceil + \cdots + \text{lg}\lceil \frac{n}{2} \rceil + \text{lg}\lceil \frac{n}{2} \rceil \\
&= \lceil \frac{n}{2} \rceil \text{lg}\lceil \frac{n}{2} \rceil\\
&\ge \frac{n}{2} \text{lg} \frac{n}{2}\\
\end{aligned}

Lorsque $n \ge 4$, on a 
$$\frac{n}{2} \ge \sqrt n $$, alors, 
\begin{aligned}
\text{lg}(n!) &\ge \frac{n}{2} \text{lg} \frac{n}{2} \\
&\ge \frac{n}{2} \text{lg} (\sqrt n)\\
&= \frac{1}{4} n \text{lg}n\\
\end{aligned}

$\exists\ n_0 =4, c = \frac{1}{4}$, tel que: $\forall n\ge n_0$, on a,
$$\text{lg}(n!) \ge c\ n\text{lg}n $$
On peut dire,
$$\text{lg}(n!) \in \Omega (n\text{lg}n)$$

**Finalement**, on peut dire,
$$\text{lg}(n!) \in \Theta (n\text{lg}n)$$
