### Question

Voici quelque réponses trouvées dans l’intra, parfois énéralisées. Dans chaque cas, expliquez ce qui ne va pas et soit donnez un exemple pour prouver vos dires, soit prouvez que l’énoncé est faux (ou vrai), suivant le cas. Parfois il y a plusieurs problèmes (et peut-être aucun?).

1. En supposant que $f,g: \mathbb N \rightarrow \mathbb R^{\ge 0}$ sont événtuellement non-décroissantes:
$$\lg \lim _{n\rightarrow \infty} \frac{f(n)}{g(n)} = 
\lim_{n\rightarrow \infty} \frac{\lg f(n)}{\lg g(n)}$$

2. $$\lim_{n\rightarrow \infty} \frac{2^{n\lg n}}{2^{n\log_3 n}} = 
\lim_{n\rightarrow \infty} \frac{n\lg n}{n\log_3 n} =
\frac{\lg n}{\lg n \log_3 2} = \lg 3$$
et donc
$$\Omega(2^{n\lg n}) = \Omega(2^{n\log_3 n})$$

3. $$\lg n! \ge \lg \lceil \frac{n}{2} \rceil + 
\lg \lceil \frac{n}{2} \rceil + \cdots +
\lg \lceil \frac{n}{2} \rceil 
\ge \frac{n}{2}\lg \lceil \frac{n}{2} \rceil
\ge \frac{1}{2} n \lg \frac{n}{2} = 
\frac{1}{2}n \lg n - \frac{n}{2} 
\ge \frac{1}{2} n \lg n$$
donc avec $c=\frac{1}{2}$ et $n_0 = 2$ on a $\lg n! \in \Omega(n\lg n)$.


4. $$\lg n! \ge \frac{n}{2} \lg\frac{n}{2} \ge \frac{n}{2}\lg \sqrt{n} = \frac{n}{4}\lg n$$
pour $n \ge 4$. Donc $\lg n! \in \Omega(n\lg n)$

### Answer

1. It's wrong. We give an counter example where $f(n) = 2^n, g(n) = 3^n$, such that:
    $$\lg \lim _{n\rightarrow \infty} \frac{f(n)}{g(n)} =
    \lg \lim _{n\rightarrow \infty} \frac{2^n}{3^n} = 
    \lg \lim _{n\rightarrow \infty} \Bigl(\frac{2}{3} \Bigr)^n = -\infty
    $$
    whereas,
    $$\lim_{n\rightarrow \infty} \frac{\lg f(n)}{\lg g(n)}=
    \lim_{n\rightarrow \infty} \frac{\lg 2^n}{\lg 3^n}=
    \lim_{n\rightarrow \infty} \frac{n}{n\lg 3}= \log_3 2
    $$

2. It's wrong. The correct way is:
    \begin{align*}
    &\lim_{n\to\infty}\frac{2^{n\lg n}}{2^{n \log_3 n}}\\
    =& \lim_{n\to\infty}\frac{2^{n\lg n}}{2^{n\lg n \log_3 2}}\\
    =&\lim_{n\to\infty} \bigl( 2^{n\lg n} \bigr) ^{1-\log_3 2} \\
    =&\lim_{n\to\infty} \bigl( 2^{1-\log_3 2} \bigr) ^{n\lg n} \\
    =& +\infty \quad (2^{1-\log_3 2} > 1)\\
    \Rightarrow &  2^{n\lg n} \in \Omega(2^{n\log_3 n}) \text{ et } 2^{n\lg n} \notin \Theta(2^{n\log_3 n})\\
    \Rightarrow & O(2^{n\lg n}) \supset O(2^{n\log_3 n})
    \end{align*} 

3. It's wrong, because to satisfy:
    $$\frac{1}{2}n \lg n - \frac{n}{2} \ge \frac{1}{2} n \lg n$$
    we need both $n\le 0$ and $n \ge 2$,  which is impossible.

    The correct way could be the following:

    Assume a constant $c$ satisfies: $$\frac{1}{2}n \lg n - \frac{n}{2} \ge c\ n \lg n$$

    we get: $$c \le \frac{1}{2} ( 1 - \frac{1}{\lg n})$$

    we can set $n_0 = 4, c = \frac{1}{4}$ so that:

    $$\forall n \ge n_0, \lg n! \ge c\ n\lg n$$

4. I think it's correct. because $\forall n \ge 4$, we always have:
    $$\frac{n}{2} \ge \sqrt{n}$$
    which proves: $\forall n \ge 4$ and $c_0 = \frac{1}{4}$, $$\lg n! \ge c_0\ n\lg n$$
    thus, $$\lg n! \in \Omega(n\lg n)$$
